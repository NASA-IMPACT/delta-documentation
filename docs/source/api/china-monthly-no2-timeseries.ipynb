```python
import rioxarray
import stackstac
from satsearch import Search
```

```python
stac_api_url = 'https://staging-stac.delta-backend.xyz/'
china_bbox = [
    73.675,
    18.198,
    135.026,
    53.459
]
datetime = "2000-01-01T00:00:00Z/2022-01-02T00:00:00Z"
collection = 'no2-monthly'

search = Search.search(
    url=stac_api_url,
    bbox=china_bbox,
    datetime=datetime,
    collections=[collection],
    limit=1000
)
items = search.items()
```

```python
len(items)
```

```python
stack = stackstac.stack(items)
```

```python
stack
```

```python
# Subset to Bounding Box for China
subset = stack.rio.clip_box(
    minx=china_bbox[0],
    miny=china_bbox[1],
    maxx=china_bbox[2],
    maxy=china_bbox[3]
)
subset
```

```python
# select the band default
data_band = subset.sel(band='cog_default')
data_band
```

```python
# Group data into months
no2_months = data_band.groupby('time')
```

```python
# Average over entire spatial bounding box for each month
monthly_mean_no2 = no2_months.mean(dim=('x', 'y'))
```

```python
%%time
monthly_mean_no2.plot()
```
